```python
from transformers import pipeline
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('test(in).csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Check data types of columns
print(df.dtypes)

# Get basic statistics of the dataset
print(df.describe())
```

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK data
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

def clean_text(text):
    # Lowercase the text
    text = text.lower()
    
    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    
    # Tokenize the text
    words = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    return ' '.join(words)

# Apply text cleaning to the 'body' column (not 'text')
df['cleaned_text'] = df['body'].apply(clean_text)

# Display the first few rows of the dataframe to verify the cleaning
df.head()
```

```python
from transformers import pipeline

# Force the model to use CPU instead of MPS
nlp = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=-1)

# Function to apply sentiment classification using LLM
def get_sentiment_llm(text):
    result = nlp(text)[0]
    return result['label']  # 'LABEL_0' = Negative, 'LABEL_1' = Positive

# Apply sentiment labeling using the pre-trained model
df['Sentiment_LLM'] = df['cleaned_text'].apply(get_sentiment_llm)

# Display the first few rows of the dataframe to verify the sentiment labeling
print(df.head())
```

```python
# Visualize sentiment distribution
sns.countplot(x='Sentiment_LLM', data=df)
plt.title('Sentiment Distribution')
plt.show()
```

```python
# Convert the 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Group by date and sentiment to count occurrences
sentiment_by_date = df.groupby([df['date'].dt.to_period('M'), 'Sentiment_LLM']).size().unstack(fill_value=0)

# Plot sentiment trends over time
sentiment_by_date.plot(kind='line', stacked=True, figsize=(10, 6))
plt.title('Sentiment Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sentiment Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize sentiment distribution by sender
sns.countplot(x='from', hue='Sentiment_LLM', data=df)
plt.title('Sentiment Distribution by Sender')
plt.xticks(rotation=90)
plt.show()
```

```python
sentiment_summary = df['Sentiment_LLM'].value_counts()
print("Sentiment Summary:\n", sentiment_summary)
```

```python
df.to_csv('sentiment_labeled_data.csv', index=False)
```
